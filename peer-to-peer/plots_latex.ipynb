```julia
using StatsPlots, Measures
```

```julia
up =  [0 31.37 31.37 31.37 31.37 31.37];
mid = [0 28.31 29.47 22.97 24.39 19.45];
low = [0 25.24 27.56 14.57  17.40 7.53];

yield_up =  [0 4.66 2.90+9.87 12.77 13.97+9.87 23.84];
yield_mid = [0 2.33 11.32 6.38 18.53 14.78];
yield_low = [0 0 9.87 0 9.87 0];
```

```julia
price = up .- low;
price = vcat(price, low)
```

```julia
yield = yield_up .- yield_low;
yield = vcat(yield, yield_low)
```

```julia
# price
groupedbar(repeat("#" .* string.(1:6)), price', bar_position=:stack, legend=false,  linealpha=0,
            color =[:teal :false], alpha=[0.6 0],
            bar_width=.2, size=(500,300), right_margin=1cm)
scatter!((2:6).-.5, mid'[2:6], markershape=:hline, markersize=11, color=[:black])
annotate!((2:6).-.1, mid[2:6], [mid[2:6]...], Plots.font("Sans", 9))
yaxis!("ct/kWh", font(8, "sans-sarif"), tickfontsize=8)
xaxis!("Cases", font(8, "sans-sarif"), tickfontsize=8)

#yield
#groupedbar!(2:4, yield'[2:4,:], bar_position=:stack, legend=false, color =[:yellowgreen :false], 
 #           linewidth=[.5 0], bar_width=.2)
scatter!((2:6).-.5, yield_mid'[2:6], markershape=:cross, markersize=11, color=[:black])
annotate!((2:6).-.1, yield_mid[2:6], [yield_mid[2:6]...], Plots.font("Sans", 9))
```
